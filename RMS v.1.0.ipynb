```python
class RMS:
    def __init__ (self,restaurant_name,restaurant_menu):
        self.user_order=''
        self.user_bill=0
        self.rest_name=restaurant_name
        self.menu=restaurant_menu
        self.user_rep=''
        self.user_pay=0
    def welcome_user(self):
        print('Welcome to',self.rest_name)
    def display_menu(self):
        print('Menu:')
        for i in self.menu:
            print(i.title(),self.menu[i])
    def take_order(self):
        self.user_order=input('Please place your order here:')
    def prepare_order(self):
        print('Preparing your',self.user_order.title())
        import time
        time.sleep(1)
        self.user_bill=self.user_bill+self.menu[self.user_order]
    def serve_order(self):
        print('Your order is ready!')
        print('Please enjoy your',self.user_order.title())
    def display_bill(self):
        print('Total Bill',self.user_bill)
    def verify_bill(self):
        self.user_pay=int(input('Please pay your bill here:'))
        while self.user_pay<self.user_bill:
            self.user_bill=self.user_bill-self.user_pay
            print('Payment Unsuccessful!')
            print('Please pay remaining',self.user_bill)
            self.user_pay=int(input('Please pay your bill here:'))
        if self.user_pay>self.user_bill:
            print('Payment Successful!')
            print('Here is your change:',self.user_pay-self.user_bill)
        else:
            pass
    def thank_user(self):
        print('Thank you for visiting',self.rest_name)
    def order_process(self):
        self.display_menu()
        self.take_order()
        if self.user_order in self.menu:
            self.prepare_order()
            self.serve_order()
            self.user_rep=input('Do you want to order anything else?')
            while self.user_rep.lower()=='yes':
                self.repeat_order()
                self.user_rep=input('Do you want to order anything else?')
            self.display_bill()
            self.verify_bill()
            self.thank_user()
        else:
            print('Invalid Order!')
            self.order_process()
    def repeat_order(self):
        self.display_menu()
        self.take_order()
        if self.user_order in self.menu:
            self.prepare_order()
            self.serve_order()
        else:
            print('Invalid Order!')
            self.repeat_order()
        
        
        
        
```

```python
if __name__=='__main__':
    rn='Sunlight Cafe'
    rm={'cold coffee':200,'latte':190,'frappe':200,'cappuccino':210}
    moonlight_cafe=RMS(rn,rm)
    import tkinter as tk
    window=tk.Tk()
    window.title('Omkar Jagtap RMS')
    window.geometry('300x300')
    tk.Label(window,text=moonlight_cafe.rest_name,font=('Helvetica',16)).place(x=75,y=70)
    tk.Button(window,text='START ORDER',command=moonlight_cafe.order_process,width=17).place(x=80,y=150)
    window.mainloop()
```
